```python
from geopy.geocoders import Nominatim
import pandas as pd

geolocator = Nominatim(user_agent="myApp")

def get_location_info(address):
    location = geolocator.geocode(address)
    if location:
        address_parts = location.raw.get("address", {})
        country = address_parts.get("country", "")
        state = address_parts.get("state", "")
        city = address_parts.get("city", "")
        return f"{address}, {city}, {state}, {country}"
    else:
        return "Tidak dapat menemukan lokasi."

# Buat DataFrame dengan nama alamat
df = pd.DataFrame({'Address': ['Kimia Farma']})

# Gunakan fungsi untuk mendapatkan nama tempat dengan informasi provinsi
asw = df['Address'].apply(get_location_info)
asw
```
